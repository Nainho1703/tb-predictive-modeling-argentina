```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Jul 21 14:20:28 2023

@author: leonardo
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# import shapefile as shp  # Requires the pyshp package
# import matplotlib.pyplot as plt
import geopandas as gpd
from fuzzywuzzy import process
import numpy as np
# ==========================================================================
# ============== Replaces %90 and more similar strings ===================== 
# ==========================================================================

# ==========================================================================
# ==========================================================================






DF1 = pd.read_excel('Bases\\base2009_2018_Leonardo_Lopez.xlsx')
DF2 = pd.read_excel('Bases\\base_2019_2022_Leonardo_Lopez_al_01_10_2023.xlsx')
DF0=pd.concat([DF1,DF2])
DF0.head()
```

```python

def replace_similar_values(value):
    # print('Reescribiendo departamentos')

    for standard_value in list(localidades):
        # print(standard_value)
        if process.extractOne(value, [standard_value])[1] > 80:
            return standard_value
    return value

DF=DF0.copy()
shape=gpd.read_file('Codgeo_Pais_x_dpto_con_datos\\pxdptodatosok.shp')

# Limpiar Bse de Datos

ProvinciasMapa=shape['provincia'].unique()
DF['PROVINCIA_RESIDENCIA']=DF['PROVINCIA_RESIDENCIA'].replace('CABA', 'Ciudad Autónoma de Buenos Aires')
DF['DEPARTAMENTO_RESIDENCIA']=DF['DEPARTAMENTO_RESIDENCIA'].astype(str)
#DF=DF.dropna()
for i in ProvinciasMapa:
    localidades=shape.loc[shape['provincia']==i]['departamen'].unique()
    print(i)
    DF.loc[DF['PROVINCIA_RESIDENCIA']==i,'DEPARTAMENTO_RESIDENCIA'] = DF.loc[DF['PROVINCIA_RESIDENCIA']==i,'DEPARTAMENTO_RESIDENCIA'].apply(replace_similar_values)


DF





    
```

```python
localidades=shape['departamen'].unique()
DF['DEPARTAMENTO_RESIDENCIA'] = DF['DEPARTAMENTO_RESIDENCIA'].apply(replace_similar_values,localidades)



DF

```
