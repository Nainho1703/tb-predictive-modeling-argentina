```python
from datetime import datetime, timedelta

# Fechas de inicio y fin
start_date = datetime.strptime('20171126', '%Y%m%d')
end_date = datetime.strptime('20240517', '%Y%m%d')

# Crear una lista para almacenar las fechas
date_list = []

# Iterar sobre el rango de fechas
current_date = start_date
while current_date <= end_date:
    date_list.append(current_date.strftime('%Y%m%d'))
    current_date += timedelta(days=1)

# URL base
base_url = "https://ssl.smn.gob.ar/dpd/descarga_opendata.php?file=observaciones/datohorario{}.txt"

# Crear una lista de URLs con las fechas
url_list = [base_url.format(date) for date in date_list]
import requests
# Imprimir las primeras 5 URLs para verificar
for date_ in date_list[:5]:



    # Formatear la URL con la fecha específica
    url_with_date = base_url.format(date_)

    # Imprimir la URL resultante
    print(url_with_date)



    # Realiza la solicitud GET
    response = requests.get(url_with_date)

    # Verifica que la solicitud fue exitosa
    if response.status_code == 200:
        # Guarda el contenido en un archivo local
        with open(date_, 'wb') as file:
            file.write(response.content)
        print('Archivo descargado con éxito')
    else:
        print(f'Error al descargar el archivo. Código de estado: {response.status_code}')
```
