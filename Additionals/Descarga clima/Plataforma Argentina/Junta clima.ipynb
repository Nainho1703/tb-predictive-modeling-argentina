```python
import os
import pandas as pd
# Especifica la ruta de la carpeta
carpeta = 'Datos clima\\'

# Obtn una lista de todos los archivos en la carpeta
archivos = os.listdir(carpeta)

# Filtra solo los archivos (excluyendo directorios)
archivos = [archivo for archivo in archivos if os.path.isfile(os.path.join(carpeta, archivo))]

# Lista los primeros cinco archivos
primeros_cinco = archivos

print(primeros_cinco)


columnas = ['FECHA', 'HORA', 'TEMP', 'HUM', 'PNM', 'DD', 'FF', 'NOMBRE']
df_clima=pd.DataFrame(columns=["AUX"])
nombre='AUX'
for f in primeros_cinco:

    df_aux = pd.read_csv("Datos clima\\"+f,delimiter="\t", names=[nombre], encoding='latin1',skiprows=2)

    df_clima=pd.concat([df_clima,df_aux])
df_clima=df_clima.reset_index(drop=True)
```

```python
df_clima

# Dividir la columna 'NOMBRE' en varias columnas
nombre_expanded = df_clima[nombre].str.split(expand=True,n=7)

# Unir las columnas originales con las nuevas columnas
df = df_clima.drop(columns=[nombre]).join(nombre_expanded)


df.columns=columnas

```

```python
df["NOMBRE"]=df["NOMBRE"].str.strip()
```

```python
col_values=["TEMP","HUM","PNM","DD","FF"]
df2=df.copy()
for c in col_values:
    print(c)


    df2[c] = pd.to_numeric(df2[c], errors='coerce')
```

```python
df_aux=pd.pivot_table(df2,index=["NOMBRE","FECHA"],values=col_values,aggfunc="mean")

df_clima_f=pd.DataFrame(df_aux.to_records())
```

```python
df_clima_f["FECHA"] = pd.to_datetime(df_clima_f["FECHA"], format='%d%m%Y')

# Convertir las fechas al formato %Y-%m-%d
df_clima_f["FECHA"] = df_clima_f["FECHA"].dt.strftime('%Y-%m-%d')
df_clima_f

df_clima_f=df_clima_f.loc[df_clima_f["TEMP"]<50]

# Get the path of the folder two levels up
two_levels_up = os.path.abspath(os.path.join(os.getcwd(), "../../"))

print(f"Path two levels up: {two_levels_up}")

```

```python
df_clima_f.to_csv(two_levels_up+"\\Bases\\Climatologicas_Argentina\\Datos_Clima_Arg.csv",index=False)
```
